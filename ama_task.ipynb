```python
#### Do products that get bigger discounts attract more reviews?"¶
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '').astype(float)
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').astype(float)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['discount_percentage'], y=df['rating_count'], alpha=0.6, color='blue')
sns.regplot(x=df['discount_percentage'], y=df['rating_count'], scatter=False, color='red')
plt.xlabel('discount rate (%)')
plt.ylabel('Number of ratings')
plt.title('Relationship between discount percentage and number of ratings')
plt.show()
```

```python
## Do top-rated products get lower or more discounts?
```

```python
plt.figure(figsize=(10, 6))
df['rating'] = df['rating'].astype(float)
sns.scatterplot(x=df['rating'].astype(float), y=df['discount_percentage'], alpha=0.6, color='green')
sns.regplot(x=df['rating'].astype(float), y=df['discount_percentage'], scatter=False, color='red')
plt.xlabel('Rating')
plt.ylabel('discount rate (%)')
plt.title('Relationship between rating and discount ratio')
plt.show()
```

```python
#### Do more expensive products get higher ratings?
```

```python
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '', regex=True).str.replace(',', '', regex=True).astype(float)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df = df.dropna(subset=['rating', 'actual_price'])
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['actual_price'], y=df['rating'], alpha=0.6, color='purple')
sns.regplot(x=df['actual_price'], y=df['rating'], scatter=False, color='red')
plt.xlabel('Actual Price (₹)')
plt.ylabel('Rating')
plt.title('Relationship between Price and Rating')
plt.show()
```

```python
### Is there a relationship between the number of ratings and the price of the product?
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df['actual_price'], y=df['rating_count'], alpha=0.6, color='orange')
sns.regplot(x=df['actual_price'], y=df['rating_count'], scatter=False, color='red')
plt.xlabel('Actual Price (₹)')
plt.ylabel('Number of ratings')
plt.title('Relationship between price and number of ratings')
plt.show()
```

```python
### Is there a relationship between the product category and the average ratings?
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
category_avg_rating = df.groupby('category')['rating'].mean().sort_values()
plt.figure(figsize=(24, 12))
category_avg_rating.plot(kind='line', color='blue')
plt.xlabel('class')
plt.ylabel('average rating')
plt.title('Average rating per product category')
plt.xticks(rotation=90)
plt.show()
```

```python
x1 = pivot_avg_rating = df.pivot_table(index='category', values='rating', aggfunc='mean')
x2 = pivot_count_products = df.pivot_table(index='category', values='product_name', aggfunc='count')
x3 = pivot_avg_price = df.pivot_table(index='category', values='actual_price', aggfunc='mean')


with pd.ExcelWriter("amzon_task.xlsx") as writer:
    df.to_excel(writer, sheet_name="main table", index=False)
    x1.to_excel(writer, sheet_name="x1", index=True)
    x2.to_excel(writer, sheet_name="x2", index=True)
    x3.to_excel(writer, sheet_name="x3", index=True)
  
print("amzon_task.xlsx' has been saved successfully!")
```
